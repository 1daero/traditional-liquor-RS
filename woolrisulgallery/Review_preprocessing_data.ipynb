```python
from konlpy.tag import Okt
from konlpy.tag import Kkma
```

```python
okt = Okt()
kkma = Kkma()

print('OKT 형태소 분석 :',okt.morphs("열심히 코딩한 당신, 연휴에는 여행을 가봐요"))
print('OKT 품사 태깅 :',okt.pos("열심히 코딩한 당신, 연휴에는 여행을 가봐요"))
print('OKT 명사 추출 :',okt.nouns("열심히 코딩한 당신, 연휴에는 여행을 가봐요")) 
```

```python
twitter = Okt() # konlpy 로 명사 추출
twitter.nouns('오미자를 좋아해서 주문했어요. 색깔은 좋은데 맛은 맛없지도 맛있지도 않은 보통맛. 그래도 다시 주문할지도. 사장님께 조언 한마디 할게요. 오미자 열매를 포도알같이 톡 터질때 넣으시면 2.4% 가 아니라 1/10만 넣어도 색상이 세계 최고가 될거에요. 오미자 소주를 담가봐서 알아요.더보기',
)
```

```python
comments = df[4] # 리뷰 데이터
comments
```

```python
comments_list= comments.values.tolist() #리스트로 변환
comments_list
```

```python
from tqdm import tqdm
```

```python
noun_list = []
for x in tqdm(comments_list) :
    a = twitter.nouns(x)
    noun_list.append(a)
    print(a)
```

```python
noun_list
```

```python
df[5] = noun_list
```
