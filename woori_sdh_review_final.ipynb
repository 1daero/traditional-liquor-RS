```python
import pandas as pd
import numpy as np
```

```python
df_sdh_review = pd.read_csv('sooldamhwa_review_pretty.csv')
df_sdh_review.drop(columns = ['Unnamed: 0','page'], axis = 1, inplace = True)
```

```python
df_sdh_review.rename(columns = {'index':'total_index', 'user':'customerID', 'comment' : 'comments'}, inplace = True)
df_sdh_review
```

```python
df_woori_review = pd.read_excel('woorisool_review_edited.xlsx')
df_woori_review
```

```python
df_woori_review = df_woori_review[['total_index', '제품명', 'customerID', 'score', 'comments']]
df_woori_review
```

```python
df_review = pd.concat([df_woori_review, df_sdh_review])
df_review
```

```python
# csv파일로 저장
df_review.to_csv('reviews.csv')
```
