```python
import pandas as pd
import re
```

```python
df_thesool = pd.read_csv('thesool.csv')
df_thesool
```

```python
# thesool 제품명 수정
name_list = []
for item in (df_thesool['제품이름']):
    new_text = re.sub(r"[^\uAC00-\uD7A30-9a-zA-Z\s]", "", item)
    name_list.append(new_text.replace(" ", ""))
df_thesool['제품명'] = name_list
df_thesool.drop('Unnamed: 0', axis = 1, inplace = True)
df_thesool.to_csv('thesool_product_name_edited.csv', index = False)
```

```python
# 제품명만 리스트 excel파일로 별도 저장
df_thesool['제품명'].to_excel('product_name_list.xlsx')
```
