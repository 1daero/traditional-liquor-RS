## rate 숫자화

```python
df_sdh_rate1 = pd.read_csv("./sooldamhwa/sooldamhwa_rate_edited.csv")
df_sdh_rate1
```

```python
# text 값들을 숫자로 바꿔준다~~~!!!
df_sdh_rate = df_sdh_rate1.replace({'text1' : {"없음" : 0, "매우약함" : 1, "약함" : 2, "중간" : 3, "매우강함" : 5, "강함" : 4},
    'text2' : {"없음" : 0, "매우약함" : 1, "약함" : 2, "중간" : 3, "매우강함" : 5, "강함" : 4},
    'text3' : {"없음" : 0, "매우약함" : 1, "약함" : 2, "중간" : 3, "매우강함" : 5, "강함" : 4},
    'text4' : {"없음" : 0, "매우약함" : 1, "약함" : 2, "중간" : 3, "매우강함" : 5, "강함" : 4},
    'text5' : {"없음" : 0, "매우약함" : 1, "약함" : 2, "중간" : 3, "매우강함" : 5, "강함" : 4}})
df_sdh_rate
```

```python
# text 컬럼들을 int로 바꿔주기 위해 한번 더 전처리
df_sdh_rate = df_sdh_rate.replace({'text5' : {"0" : 0}})
df_sdh_rate.info()
```
