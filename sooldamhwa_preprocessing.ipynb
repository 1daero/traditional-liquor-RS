```python
### 컬럼 합치기
```

```python
# 먼저 text 컬럼들을 str으로 변환 -> astype 함수

df_rate_total[['text1','text2','text3','text4','text5']] = df_rate_total[['text1','text2','text3','text4','text5']].astype(str)
df_rate_total
```

```python
df_rate = pd.DataFrame()
df_rate['index'] = df_rate_total['index']
df_rate['제품명'] = df_rate_total['제품명']
df_rate['page'] = df_rate_total['page']
df_rate
```

```python
for i in range(1,6) : 
    df_rate[f'rate{i}'] = df_rate_total[f'label{i}'] + "=" + df_rate_total[f'text{i}']
df_rate
```

```python
### melt
```

```python
df_rate = df_rate.melt(id_vars=['index','제품명','page']).sort_values(['index','variable']).reset_index(drop=True)
df_rate
```

```python
### seperate
```

```python
df_rate['col'] = df_rate['value'].str.split('=').str[0]
df_rate['rate'] = df_rate['value'].str.split('=').str[1]
df_rate
```

```python
df_rate.drop(['variable','value'],axis=1, inplace=True)
df_rate
```

```python
### pivot
```

```python
df_rate_wide=pd.pivot(df_rate, index=['index','제품명','page'], columns = 'col',values = 'rate')
df_rate_wide
```

```python
df_rate_wide.drop(['0','고소한 향','단맛','맛의 강도','바디감','산미','화사한 향'],axis=1,inplace=True)
```
