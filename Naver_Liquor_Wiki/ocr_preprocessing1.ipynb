# OCR 결과 csv 파일 전처리

```python
# 필요패키지 import
import time
import json
import urllib
import numpy as np
import pandas as pd
import csv
from collections import Counter
```

### ocr name_feature 리스트로 불러와서 중복값을 제거

```python
name_list = []
f = open('./name_feature.csv', 'r', encoding= 'utf-8')
reader = csv.reader(f)
for row in reader:
    name_list.append(row)
f.close()  
```

```python
name_list
```

```python
# header가 함께 불러와졌기 때문에 삭제
del name_list[0]
# 중복값 제거 : dict.fromkeys를 이용해 순서를 지키며 중복값을 없애줌!
name_list = list(map(lambda x: list(dict.fromkeys(x)), name_list))
```

```python
# 리스트 내 value count를 위해 1차원 리스트로 변환
name_list1 = sum(name_list, [])
# list 내 value count -> counter 사용
result = Counter(name_list1)
print(result)
```

```python
# value count를 바탕으로 feature 추출 (수작업...)
feature_list = ['바디감','신맛','단맛','청량감','목넘김','맛','향기','향']
```

```python
new_name_list = []

for i in range(len(name_list)) :
    new_name_list.append(list(filter(lambda x : x in feature_list, name_list[i])))
new_name_list
```

```python
# df로 변환!
df_name = pd.DataFrame(new_name_list)
df_name
```
