```python
import pandas as pd
import numpy as np
```

```python
a = pd.read_excel('labled_liquor_list.xlsx')
a
```

```python
df_naverlist = pd.DataFrame(data = np.array(a['제품명(띄어쓰기삭제)']).reshape(-1,7),
                            columns = ['제품명', 'b', 'c', 'd', 'e', 'f', 'g'])
df_naverlist = df_naverlist.drop(['b', 'c', 'd', 'e', 'f', 'g'], axis = 1, inplace = False)
df_naverlist
```

```python
df_feature = pd.DataFrame(data = np.array(a['점수']).reshape(-1,7),
                          columns = ['향기', '단맛', '신맛', '청량감', '바디감', '목넘김', '균형감'])
df_feature
```

```python
df_naver_wiki = pd.concat([df_naverlist,df_feature], axis=1 )
df_naver_wiki
```

```python
df_naver_wiki.to_csv('naver_wiki_total.csv')
df_naver_wiki.to_excel('naver_wiki_total.xlsx')
```
