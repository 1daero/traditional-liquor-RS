# 술담화 리뷰 preprocessing

merge / index 부여 / 결측치 제거

```python
import numpy as np
import pandas as pd
import csv
import konlpy
```

### review csv merge

```python
df_review1 = pd.read_csv("./sooldamhwa_review.csv")
df_review2 = pd.read_csv("./sooldamhwa_review2.csv")
```

```python
df_review3 = pd.concat([df_review1, df_review2], axis = 0)
df_review3
```

### page 기준으로 index 부여

```python
df_sdh_rate1 = pd.read_csv("./sooldamhwa_rate_edited.csv")
df_sdh_rate1
```

```python
df_total = pd.read_excel("../Naver_Liquor_Wiki/naver_wiki_total.xlsx")
df_total1 = pd.merge(df_total, df_sdh_rate1, left_on='index', right_on='total_index', how='inner')
df_total1.columns
```

```python
df_total1.drop(['향기', '단맛', '신맛', '청량감', '바디감', '목넘김', '균형감','total_index', 'name', 'label1', 'label2', 'label3', 'label4', 'label5',
       'text1', 'text2', 'text3', 'text4', 'text5' ],axis=1, inplace=True)
df_total1
```

```python
# sooldamhwa index list 저장...

df_total1.to_csv('./sooldamhwa_index_list.csv', index = False)
```

```python
### 결측치 제거
```

```python
# index list에 값이 있는 review만 남기기
df_review = pd.merge(df_total1, df_review3, left_on = 'page', right_on ='product_id', how='left')
df_review
```

```python
df_review.info()
```

### 결측치 처리

```python
# product_id에 null 값 있는 행 출력
df_review[df_review.product_id.isnull()]
```

```python
# review가 없는 제품의 경우 null로 출력! -> 제거.

df_review.dropna(subset = ['product_id'],inplace= True)
df_review
```

```python
# product id, name 컬럼 제거
df_review.drop(['product_name', 'product_id'],axis=1, inplace=True)
```

```python
# comment가 없는 컬럼의 경우 Nan -> 0 으로 대치

df_review.loc[df_review['comment'] != df_review['comment'], 'comment'] = 0
```
