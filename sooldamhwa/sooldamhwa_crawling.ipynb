food_wrapper만 불러와서 list에 저장, df 만들기

```python
# 중간중간 비어있는 페이지, 술 단일 품목이 아닌 세트 구성 상품이 많기 때문에 이를 제외하고 돌아가도록 if문 설계



foodwrapper_list = []

for page in range(1, 370) :
    #time sleep   
    seed = np.random.randint(100)
    np.random.seed(seed)
    a = np.random.randint(5) 
    time.sleep(a)
    # url 설정
    url = f"https://www.sooldamhwa.com/damhwaMarket/detail/{page}"
        # requests로 데이터 요청하기
    resp = requests.get(url)
    html = BeautifulSoup(resp.text, "html.parser")

    if resp.status_code == requests.codes.NOT_FOUND:
        print(f"{page}페이지를 건너뜁니다.")

    elif not html.select("div.food-wrapper") :
        print(f"{page}페이지는 술 제품이 아니네요!")

    elif resp.status_code == requests.codes.ok:
        print(f"{page}페이지 크롤링 중입니다.")
        # html로 변환
        html = BeautifulSoup(resp.text, "html.parser")
        # page 숫자 추가
        foodwrapper_list.append(f"{page}")
        # 술 이름 추가
        liquorname = html.select("div.productName")[0].text
        foodwrapper_list.append(liquorname)
        # 데이터 선별
        for food in html.select("div.food-wrapper")[:5] :
            foodwrapper_list.append(food.text)

         
print("크롤링 완료!")
```

```python
foodwrapper_list = np.array(foodwrapper_list).reshape(-1,5)
foodwrapper_list
```

```python
df_sdh_food = pd.DataFrame(data = foodwrapper_list, columns= ["page","name","food1","food2","food3"])
df_sdh_food
```

```python
# csv 파일로 내보내기

df_sdh_food.to_csv("sooldamhwa_food.csv", index= False)
```

```python
ratestyle_list = []

for page in page_list :
    #time sleep   
    seed = np.random.randint(100)
    np.random.seed(seed)
    a = np.random.randint(3) 
    time.sleep(a)
    # url 설정
    url = f"https://www.sooldamhwa.com/damhwaMarket/detail/{page}"
        # requests로 데이터 요청하기
    resp = requests.get(url)
    html = BeautifulSoup(resp.text, "html.parser")

    if not html.select("div.DetailSoolRatesStyle__Label-sc-1t255cq-2") :
        print(f"{page}페이지 오류!")

    elif resp.status_code == requests.codes.ok:
        print(f"{page}페이지 크롤링 중입니다.")
        # page 숫자 추가
        ratestyle_list.append(f"{page}")
        # 술 이름 추가
        liquorname = html.select("div.productName")[0].text
        ratestyle_list.append(liquorname)
        # 라벨 데이터 선별
        for label in html.select("div.DetailSoolRatesStyle__Label-sc-1t255cq-2")[:5] :
            ratestyle_list.append(label.text)
        # 텍스트 데이터 선별
        for t in html.select("div.DetailSoolRatesStyle__Text-sc-1t255cq-1")[:5] :
            ratestyle_list.append(t.text)

         
print("크롤링 완료!")
```

### dataframe 생성

```python
# 사용할 컬럼만 살려서 빈 데이터 프레임 만들기 -> 컬럼 4개
# 컬럼 순서가 뒤죽박죽으로 들어오지만, 중요하지 않아 그대로 진행하기로

df_sdh_review = pd.DataFrame(columns = {'product_id','product_name','score','comment','user'})
```

```python
header = {     
          'user-agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/107.0.0.0 Safari/537.36'
         }
         
for page in page_list :
    #time sleep   
    seed = np.random.randint(100)
    np.random.seed(seed)
    a = np.random.randint(3) 
    time.sleep(a)
    # url 설정
    url = f"https://api.sooldamhwa.com/product/review/{page}?limit=350&order=latest"
    # requests로 데이터 요청하기
    resp = requests.get(url,headers= header)
    data = json.loads(resp.text)
    
    if not data['result'] : 
        # 리뷰가 없는 페이지 지나치기
        print(f"{page}페이지 리뷰 없음")
    else :
        df = pd.DataFrame(data['result'])
        # 쓰지 않는 column 드랍
        df.drop(columns={'id','type','maker_id','product_option_id','product_option_name','product_thumbnail','img_urls','flavor_score','design_score','delivery_score','with_who','to_who','step','status','written_at','created_at'}, inplace=True)
        df_sdh_review = df_sdh_review.append(df, ignore_index = True)
        print(f"{page} 페이지 리뷰 추가!")

         
print("크롤링 완료!")
```
