```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
review = pd.read_excel('review_keyword_final.xlsx')
```

```python
review = review[['total_index', 'comments_6']]
```

```python
review
```

```python
drop_df = review[review['comments_6'].str.len() == 2]
```

```python
review_df = pd.merge(review, drop_df, how='outer', indicator=True)
```

```python
review_df2 = review_df[review_df['_merge'] == 'left_only']
```

```python
review_df2
```

```python
total_index = review.total_index.unique().tolist()
```

```python
total_index
```

```python
real_lis = []
dic = {}
for i in total_index:
    dic[i] = review_df2[review_df2['total_index'] == i]['comments_6'].values.tolist()
    real_lis.append(dic)
    dic = {}
    
    
```

```python
real_df = pd.DataFrame(columns=['total_index', 'review'])
for i in real_lis:
    df = pd.DataFrame(list(i.items()))
    real_df = pd.concat([real_df, df])
```

```python
real_df = real_df.drop(['total_index', 'review'], axis = 1)
real_df.columns = ['total_index', 'review']
```

```python
real_df = real_df.astype({'total_index':'int'})
real_df.reset_index(drop=True, inplace=True)
```

```python
real_df
```

```python
def cha(x):
    a = " ".join(x)
    return a
```

```python
real_df['review'] = real_df['review'].apply(lambda x: cha(x))
```

```python
import re
def gsub(x):
    x = re.sub('[-=,+#/\?:^$.@*\"※~&%ㆍ!』\\‘|\(\)\[\]\<\>`\'…》]','', x)
    a = x
    return a
```

```python
real_df['review2'] = real_df['review'].apply(lambda x: gsub(x))
```

```python
def spliter(x):
    a = x.split(' ')
    a = list(set(a))
    return a
```

```python
real_df['review2'] = real_df['review2'].apply(lambda x: spliter(x))
```

```python
real_df['review2'] = real_df['review2'].apply(lambda x: cha(x))
```

```python
real_df.to_csv('review_keyword.csv', index=False)
```
